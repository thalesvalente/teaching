## Slide 7 - Gramática Livre de Contexto (GLC)

Uma **Gramática Livre de Contexto (GLC)** $G$ é definida pela quádrupla:
$$G = (V, T, P, S)$$
onde:
- $V$ é o conjunto de variáveis (não-terminais),
- $T$ é o conjunto de símbolos terminais,
- $P$ é o conjunto de regras de produção da forma $A \to \alpha$, com $A \in V$ e $\alpha \in (V \cup T)^*$,
- $S$ é o símbolo inicial da gramática.

A gramática é denominada "livre de contexto" porque as regras de produção permitem que o não-terminal $A$ seja substituído por $\alpha$ independentemente do contexto ao redor de $A$.
- Gramáticas Livres de Contexto são poderosas para descrever a sintaxe de linguagens de programação e estruturas aninhadas, como parênteses balanceados e tags HTML/XML correspondentes.
- Elas são mais expressivas que gramáticas regulares e podem gerar linguagens que não são regulares, como $\{a^n b^n \mid n \geq 0\}$ (sequências de 'a's seguidas por 'b's em número igual).


### Exemplos 1


1. **GLC para palavras com parênteses balanceados:**
   - $S \to SS$
   - $S \to (S)$
   - $S \to \epsilon$
   - Esta gramática gera todas as strings de parênteses corretamente balanceados, como "", "()", "(())", etc.

**Utilização em Análise Sintática**  
- Em compiladores, GLCs são usadas para definir a gramática de uma linguagem de programação.
- A análise sintática, que verifica se uma string pode ser gerada por uma gramática, utiliza algoritmos baseados em GLC para analisar e converter código fonte em uma estrutura de dados compreensível pelo computador (geralmente uma árvore sintática).

**Importância dessa gramática em computação:**
- Esta gramática é crucial porque define uma estrutura de **duplo balanceamento**, essencial em linguagens de programação bloco-estruturadas e no manejo de parênteses balanceados. Ela exemplifica como as linguagens podem ser definidas para garantir uma correspondência correta entre elementos simétricos, fundamental na análise sintática de linguagens de programação.

**Discussão**

A relevância dessa GLC se estende à compreensão de como os compiladores verificam a correta estruturação do código, ajudando a prevenir erros de sintaxe que poderiam levar a falhas de execução ou comportamentos inesperados dos programas. Além disso, as GLCs são utilizadas para modelar muitos aspectos das linguagens naturais em processamento de linguagem natural (PLN).

### Exemplos 2

Considere a linguagem $L_1 = \{ a^n b^n | n \geq 0 \}$. A Gramática Livre de Contexto $G_1$ que gera $L_1$ é definida por:
$$ G_1 = (\{ S \}, \{ a, b \}, \{ S \to aSb | S \to \epsilon \}, S) $$

**Derivação de um exemplo:**
- Para gerar a palavra "aabb", seguimos a seguinte derivação:
  $$ S \implies aSb \implies aaSbb \implies aa\epsilon bb \implies aabb $$


**Exemplo 2.1: Geração da palavra "ab"**

- **Gramática:** $G_1 = (\{ S \}, \{ a, b \}, \{ S \to aSb | S \to \epsilon \}, S)$
- **Objetivo:** Gerar a palavra "ab"
- **Derivação:**
  - Início: $S$
  - Aplicação da regra $S \to aSb$: $aSb$
  - Aplicação da regra $S \to \epsilon$: $a\epsilon b = ab$

Este processo mostra como a palavra "ab" é construída, começando do símbolo inicial $S$ e aplicando regras de substituição para chegar à palavra final.

**Exemplo 2.2: Geração da palavra "aabb"**

- **Objetivo:** Gerar a palavra "aabb"
- **Derivação:**
  - Início: $S$
  - Aplicação da regra $S \to aSb$: $aSb$
  - Aplicação da regra $S \to aSb$ novamente: $aaSbb$
  - Aplicação da regra $S \to \epsilon$: $aa\epsilon bb = aabb$

**Exemplo 2.3: Geração da palavra "aaabbb"**

- **Objetivo:** Gerar a palavra "aaabbb"
- **Derivação:**
  - Início: $S$
  - Aplicação da regra $S \to aSb$: $aSb$
  - Aplicação da regra $S \to aSb$ novamente: $aaSbb$
  - Aplicação da regra $S \to aSb$ mais uma vez: $aaaSbbb$
  - Aplicação da regra $S \to \epsilon$: $aaa\epsilon bbb = aaabbb$


### Exemplos 3

Considere a linguagem $\boldsymbol{L_2}$, que compreende expressões aritméticas contendo colchetes balanceados, dois operadores, e um operando. A Gramática Livre de Contexto $\boldsymbol{G_2}$ que gera $L_2$ é definida por:

- **Variáveis (Não-terminais):** $V = \{E\}$
- **Terminais:** $T = \{+, *, [, ], x\}$
- **Produções ($P_2$):**
  $$ P_2 = \{
  E \to E+E, \quad
  E \to E*E, \quad
  E \to [E], \quad
  E \to x
  \}$$  
  ou ainda:
    $$ P_2 = \{  E \to E+E | E*E | [E] | x \}$$
- **Símbolo Inicial:** $S = E$  

Essa gramática permite a construção de expressões aritméticas onde operadores e operandos estão estruturados de maneira a respeitar os colchetes como delimitadores de subexpressões.

**Derivação de um Exemplo: $[x+x]*x$**

A palavra "$[x+x]*x$" pode ser gerada pela seguinte sequência de derivações:
- Início: $E$
- Aplicação da regra $E \to E*E$: 
  - $E*E$
- Substituição do primeiro $E$ por $[E]$: 
  - $[E]*E$
- Expansão de $[E]$ para $[E+E]$: 
  - $[E+E]*E$
- Substituição do primeiro $E$ dentro dos colchetes por $x$: 
  - $[x+E]*E$
- Substituição do segundo $E$ dentro dos colchetes por $x$: 
  - $[x+x]*E$
- Substituição do último $E$ por $x$: 
  - $[x+x]*x$

**Discussão: Unicidade da Sequência de Derivação**

A questão de se a sequência de derivação é única para uma dada palavra em uma GLC levanta a discussão sobre a **ambiguidade** da gramática. Uma gramática é ambígua se uma palavra pode ser derivada de mais de uma maneira. No caso da gramática $G_2$:

- Dada a estrutura das produções, múltiplas derivações podem ser possíveis para uma mesma expressão aritmética, especialmente devido à presença de múltiplos operadores e a possibilidade de inserção de expressões entre colchetes.

**Exemplo de possível ambiguidade:**

Considere a palavra "$x+x*x$":
- Pode ser derivada como $(x+x)*x$ ou $x+(x*x)$ dependendo da aplicação das regras $E \to E+E$ e $E \to E*E$, evidenciando a ambiguidade na gramática.
- Assim, a gramática $\boldsymbol{G_2}$ é potencialmente ambígua. A mesma palavra pode ser derivada de diferentes maneiras dependendo da ordem de aplicação das regras de produção, especialmente em expressões com múltiplos operadores e operandos. Isso pode ser problemático em contextos como a análise de linguagens de programação, onde a semântica precisa ser claramente definida e unívoca.

Essa discussão é fundamental para entender a importância de uma definição clara e precisa de gramáticas, especialmente na concepção de linguagens de programação e em compiladores, onde a ambiguidade pode levar a interpretações errôneas de código.

## Slide 12 - Gramática Livre de Contexto (GLC)

Uma **Árvore de Derivação** oferece uma maneira visual e estruturada de representar o processo de derivação de uma palavra em uma gramática livre de contexto. É particularmente útil para entender como uma palavra é construída a partir das regras de produção de uma gramática. As características de uma árvore de derivação incluem:

1. **Raiz:** 
    - A raiz da árvore é sempre o símbolo inicial da gramática.
2. **Vértices não-folhas:** 
    - Estes são os não-terminais. 
    - Cada nó não-folha $A$ com filhos $X_1, X_2, ..., X_n$ representa a aplicação da regra de produção $A \to X_1 X_2 ... X_n$.
3. **Vértices folha:** 
    - Estes são terminais ou o símbolo $\epsilon$. 
    - Eles representam os símbolos terminais da palavra final ou a palavra vazia, indicando o fim de uma cadeia de derivações para aquele ramo.

### Exemplo 1


Consideremos a gramática $G_2$ para expressões aritméticas com colchetes balanceados:

- **Gramática $G_2$:** $G_2 = (\{ E\},\{+,*,[,],x \},\{E\to E+E | E*E | [E] | x \},E)$

````python
from graphviz import Digraph

def create_derivation_trees():
    dot = Digraph('Derivation Trees', filename='derivation_trees', format='png')
    dot.attr(rankdir='TB', size='8,5')

    # Subgrafo para a árvore de derivação do duplo desbalanceamento
    with dot.subgraph(name='cluster_0') as c:
        c.attr(style='invisible')
        c.node('S0', 'S')
        c.node('a1', 'a', shape='none')
        c.node('S1', 'S')
        c.node('b1', 'b', shape='none')
        c.node('a2', 'a', shape='none')
        c.node('S2', 'S')
        c.node('b2', 'b', shape='none')
        c.node('S3', 'ε', shape='none')
        c.edges([('S0', 'a1'), ('S0', 'S1'), ('S0', 'b1')])
        c.edges([('S1', 'a2'), ('S1', 'S2'), ('S1', 'b2')])
        c.edges([('S2', 'S3')])
        c.attr(label='Duplo desbalanceamento')

    # Subgrafo para a árvore de derivação da expressão aritmética
    with dot.subgraph(name='cluster_1') as c:
        c.attr(style='invisible')
        c.node('E0', 'E')

        c.node('E1', 'E')
        c.node('plus', '*', shape='none')
        c.node('E2', 'E')

        c.node('x0', 'x', shape='none')

        c.node('open', '[', shape='none')
        c.node('E3', 'E')
        c.node('close', ']', shape='none')

        c.node('E4', 'E')
        c.node('add', '+', shape='none')
        c.node('E5', 'E')
        
        c.node('x1', 'x', shape='none')
        c.node('x2', 'x', shape='none')

        c.edges([('E0', 'E1'), ('E0', 'E2'), ('E0', 'plus')])
        c.edges([('E1', 'open'), ('E1', 'E3'), ('E1', 'close')])
        c.edges([('E2', 'x0')])
        c.edges([('E3', 'E4'), ('E3', 'add'), ('E3', 'E5')])
        c.edges([('E4', 'x1')])
        c.edges([('E5', 'x2')])
        c.attr(label='Expressão aritmética')

    print(dot.source)
    dot.render('derivation_trees_output', view=True)

create_derivation_trees()
````

**Interpretação da expansão da Expressão Aritmética**
- A raiz é o símbolo inicial `E`.
- O primeiro `E` se expande usando a produção `E -> E*E`, criando dois novos nós `E`.
- O primeiro `E` à esquerda do `*` expande para `[E]`, e então para `[E+E]` dentro dos colchetes.
- Dentro dos colchetes, `E` se expande para `x` e depois `x + E`, com o último `E` também se tornando `x`.
- O segundo `E` à direita do `*` expande diretamente para `x`.

**Interpretação do Duplo Desbalanceamento:**
- A **raiz** é o símbolo inicial `S`.
- O símbolo `S` se expande usando a produção `S -> aSb`, introduzindo terminais `a` e `b` e um novo não-terminal `S` entre eles.
- Este novo `S` segue a mesma regra de produção `S -> aSb` repetidamente, criando uma estrutura aninhada onde cada novo `S` está encapsulado por `a` e `b`.
- No nível mais interno, `S` utiliza a produção `S -> ε` (épsilon), que é a produção que gera a cadeia vazia, concluindo a expansão dos não-terminais.
- Cada expansão de `S` em `aSb` adiciona um par `a` e `b`, garantindo que para cada `a` inicial haja um `b` correspondente no final, resultando em palavras da forma `a^n b^n`.


### Exemplo 2

**Árvore de Derivação para $x + x \times x$:**

- A **raiz** é o símbolo inicial `E`.
- O nó raiz `E` se expande em `E + E`, separando a expressão em duas partes.
- A esquerda `E` da adição se resolve diretamente para `x`.
- O `E` à direita do `+` se expande para $E \times E$, representando a operação de multiplicação.
- Cada `E` na expressão de multiplicação resolve para `x`, completando a árvore.

**Vários Caminhos de Derivação:**

1. **Derivação direta:**
   - $ E \rightarrow E + E \rightarrow x + E \rightarrow x + (E \times E) \rightarrow x + (x \times x) $

2. **Derivação com expansão antecipada da multiplicação:**
   - $ E \rightarrow E + E \rightarrow E + (E \times E) \rightarrow E + (x \times E) \rightarrow E + (x \times x) \rightarrow x + (x \times x) $

3. **Início com multiplicação:**
   - $ E \rightarrow E + E \rightarrow (E \times E) + E \rightarrow (x \times E) + E \rightarrow (x \times x) + E \rightarrow (x \times x) + x $

Cada caminho ilustra como diferentes derivações podem levar ao mesmo resultado final, mostrando a não-determinística natureza das gramáticas livres de contexto ao derivar expressões.  
Agora, vamos gerar o código Python para desenhar a árvore de derivação da expressão $x + x \times x$


````python
from graphviz import Digraph

# Criação da árvore de derivação
dot = Digraph('Derivation Trees', filename='derivation_trees', format='png')
dot.attr(rankdir='TB', size='8,5')
dot.node('E0', 'E')
dot.node('E1', 'E')
dot.node('plus', '+', shape='none')
dot.node('E2', 'E')
dot.node('E3', 'E')
dot.node('times', '*', shape='none')
dot.node('E4', 'E')
dot.node('x1', 'x', shape='none')
dot.node('x2', 'x', shape='none')
dot.node('x3', 'x', shape='none')

# Construindo a árvore
dot.edge('E0', 'E1')
dot.edge('E0', 'plus')
dot.edge('E0', 'E2')
dot.edge('E2', 'E3')
dot.edge('E2', 'times')
dot.edge('E2', 'E4')
dot.edge('E1', 'x1')
dot.edge('E3', 'x2')
dot.edge('E4', 'x3')

# Visualização da árvore
print(dot.source)
dot.render('derivation_tree', view=True)
````

## Slide 17 - Árvore de Derivação e Tipos de Derivação**

### Derivação Mais à Esquerda**

**Definição:** Uma derivação mais à esquerda em uma gramática livre de contexto ocorre quando, a cada passo da derivação, a substituição (ou produção) é aplicada ao não-terminal mais à esquerda ainda não substituído na cadeia.

- **Exemplo Prático:**
  - Considere a palavra $x + x \times x$ e a gramática $G = (\{E\}, \{x, +, \times\}, P, E)$ onde $P$ inclui as regras:
    - $E \to E + E$
    - $E \to E \times E$
    - $E \to x$
  - **Desenvolvimento**
    - Começamos com $E$.
    - $E \implies E + E$ (substituímos o $E$ mais à esquerda por $E + E$)
    - $E + E \implies E + E$ (substituímos o $E$ mais à esquerda por $x$)
    - $x + E \implies x + E \times E$ (substituímos o $E$ mais à esquerda por $E \times E$)
    - $x + E \times E \implies x + x \times E$ (substituímos o $E$ mais à esquerda por $x$)
    - $x + x \times E \implies x + x \times x$ (substituímos o $E$ restante por $x$)

### Derivação Mais à Direita

**Definição:** Uma derivação mais à direita em uma gramática livre de contexto ocorre quando, a cada passo da derivação, a substituição é aplicada ao não-terminal mais à direita ainda não substituído na cadeia.

- **Exemplo Prático:**
  - Usando a mesma palavra e gramática do exemplo anterior.
  - **Desenvolvimento**
    - Começamos com $E$.
    - $E \implies E + E$ (substituímos o $E$ mais à direita por $E + E$)
    - $E + E \implies E + E \times E$ (substituímos o $E$ mais à direita por $E \times E$)
    - $E + E \times E \implies E + E \times x$ (substituímos o $E$ mais à direita por $x$)
    - $E + E \times x \implies E + x \times x$ (substituímos o $E$ mais à direita por $x$)
    - $E + x \times x \implies x + x \times x$ (substituímos o $E$ restante por $x$)

## Slide 18 - Ambiguidade

**Definição de Ambiguidade:**
- Uma gramática livre de contexto é considerada ambígua se uma mesma palavra pode ser gerada por mais de uma árvore de derivação. Isso significa que existem múltiplas maneiras de aplicar as regras da gramática para obter a mesma sequência de símbolos.

**Exemplo de Ambiguidade:**
- Considere a palavra $ x+x \times x $ e a gramática $ G = (\{E\}, \{x, +, \times\}, P, E) $ com as regras:
  - $ E \to E + E $
  - $ E \to E \times E $
  - $ E \to x $

- Esta palavra pode ser gerada por duas árvores de derivação distintas, ilustrando a ambiguidade na gramática. As duas derivações possíveis são mostradas abaixo:

1. Árvore de Derivação para Derivação Mais à Esquerda
    - Primeira árvore (operação de adição feita por último):
      - $ E $
      - $ \to E + E $
      - $ \to (E \times E) + E $
      - $ \to (x \times x) + x $
  
2. Árvore de Derivação para Derivação Mais à Direita
    - Segunda árvore (operação de multiplicação feita por último):
      - $ E $
      - $ \to E + E $
      - $ \to E + (E \times E) $
      - $ \to x + (x \times x) $

**Visualização das Árvores de Derivação:**
- As duas árvores de derivação apresentadas visualmente demonstram como as regras podem ser aplicadas de maneiras diferentes para alcançar o mesmo resultado final, o que caracteriza a ambiguidade da gramática.

**Impacto da Ambiguidade:**
  - A ambiguidade de uma gramática pode complicar a análise sintática automatizada
  - diferentes análises podem levar a interpretações distintas da mesma entrada. 
  - É crucial, especialmente em contextos de programação e processamento de linguagens, garantir que as gramáticas sejam inequívocas para evitar ambiguidades que possam resultar em comportamentos inesperados ou erros de interpretação.

**Conclusão:**
- A existência de múltiplas árvores de derivação para uma única palavra é um indicativo claro de ambiguidade
- Nestes casos, esforços para desambiguar gramáticas são fundamentais na concepção de compiladores e intérpretes mais eficazes e confiáveis.


## Slide 36 - Autômato com Pilha

**Definição e Uso:**
- **Autômatos com Pilha** são modelos computacionais que podem reconhecer as **Linguagens Livres de Contexto**. 
- Estes autômatos são equipados com uma memória auxiliar na forma de uma pilha, permitindo-lhes armazenar uma quantidade ilimitada de informação.
  
**Características Principais:**
- **Pilha**: Serve como memória auxiliar do autômato. Permite operações de empilhar (push), que adiciona um símbolo ao topo da pilha, e desempilhar (pop), que remove o símbolo do topo.
- **Não-determinismo**: Um autômato com pilha pode ter mais de uma transição possível para um mesmo estado e símbolo de entrada, ou pode mover-se sem consumir nenhum símbolo de entrada (transição $\epsilon$).

**Visualização da Pilha em Operação:**
- A imagem a seguir mostra o conceito de uma pilha utilizada em autômatos com pilha, destacando como os símbolos são adicionados e removidos do topo da pilha, e indicando o sentido de crescimento da pilha.

![Pilha em Autômato com Pilha](pilha.png)

**Importância do Não-Determinismo:**
- O não-determinismo é crucial porque permite que o autômato explore várias possibilidades de processamento simultaneamente, o que é particularmente útil para processar linguagens que requerem mais poder computacional do que aquele oferecido pelos autômatos finitos.

**Definição Formal:**
- Um APN é formalmente definido como uma 6-tupla:
  $ M = (\Sigma, Q, \delta, q_0, F, V) $
  onde:
  - $\Sigma$: Alfabeto de entrada.
  - $Q$: Conjunto finito de estados.
  - $\delta$: Função de transição, definida como:
    $$ \delta: Q \times (\Sigma \cup \{\epsilon\}) \times (\Gamma \cup \{\epsilon\}) \rightarrow 2^{Q \times \Gamma^*} $$ 
  - $q_0$: Estado inicial, com $q_0 \in Q$.
  - $F$: Conjunto de estados finais, com $F \subseteq Q$.
  - $V$: Alfabeto da pilha.

**Componentes da Função de Transição**

- **$\delta$**: Representa a própria função de transição do autômato.
- **$Q$**: Conjunto de estados do autômato, onde cada estado é considerado na função de transição.
- **$\Sigma \cup \{\epsilon\}$**: Representa o alfabeto de entrada do autômato unido com o símbolo $\epsilon$, permitindo transições que não consomem um símbolo de entrada.
- **$\Gamma \cup \{\epsilon\}$**: Representa o alfabeto da pilha, também unido com $\epsilon$, permitindo operações que não modificam a pilha.
- **$2^{Q \times \Gamma^*}$**: O conjunto de potências das combinações de estados e sequências de pilha, indicando todas as possíveis respostas da função de transição.

**Explicação Detalhada**

- A função $\delta$ responde com base:
  - no estado atual, 
  - um símbolo de entrada (que pode ser omitido como $\epsilon$), 
  - e um símbolo de pilha (que também pode ser $\epsilon$).
- Esta função determina as possíveis transições para outros estados e as possíveis manipulações na pilha, que podem envolver adicionar novos símbolos à pilha, remover símbolos, ou deixar a pilha inalterada.
- O não-determinismo é uma característica chave dos Autômatos com Pilha, permitindo múltiplas transições possíveis para uma dada entrada e condição de pilha.
- Essa função é crucial para o processamento de linguagens livres de contexto, onde a capacidade de lembrar e manipular sequências de símbolos através da pilha permite o reconhecimento de padrões complexos e estruturados.

**Funcionamento:**
- O autômato lê símbolos de entrada e utiliza uma pilha para armazenar símbolos que ajudam a controlar o processo de aceitação de palavras.
- As transições podem depender do 
  - estado atual
  - do símbolo de entrada (ou de um $\epsilon$-movimento) e 
  - do topo da pilha
- As transições podem resultar em mudanças de estado e operações de empilhar ou desempilhar símbolos na pilha.

Graficamente, a função programa pode ser representada como:  
![Gráfico de Autômato com Pilha](funcao_grafico.png)


**Aspectos Importantes da Função $\delta$:**
- **Totalidade:** A função $\delta$ pode ser total, cobrindo todas as combinações possíveis de estado atual, símbolo de entrada (ou $\epsilon$), e símbolo no topo da pilha.
- **Não-determinismo:** 
  - O uso do símbolo $\epsilon$ para a entrada ou a pilha introduz não-determinismo:
    - **Na entrada:** Permite transições sem consumir um símbolo da entrada.
    - **Na pilha:** Permite transições sem necessariamente depender do símbolo no topo da pilha.
- **Operações na Pilha:** 
  - Um $\epsilon$ na operação de pilha indica que não há modificação na pilha (não empilha nem desempilha nada).
- **Uso do Símbolo $?$:**
  - O símbolo $?$ tem interpretações específicas dependendo de onde é usado:
    - **Na leitura da fita:** Pode indicar se toda a entrada foi consumida, ou seja, se a leitura da fita chegou ao fim.
    - **Na leitura da pilha:** Pode ser usado para verificar se a pilha está vazia.

**Exemplo de Uso em Análise de Linguagem:**
- Autômatos com pilha são amplamente utilizados na análise sintática em compiladores, onde a precisão na manipulação da pilha e a capacidade de lidar com entradas complexas são essenciais.

### Exemplo 1 de Autômato com Pilha (AP)

Considere um Autômato com Pilha que reconhece a linguagem $L = \{a^n b^n | n \geq 1\}$, onde o número de 'a's é igual ao número de 'b's.

**Definição do Autômato com Pilha**

- **Alfabeto de Entrada $\Sigma$:** $\{a, b\}$
- **Alfabeto da Pilha $\Gamma$:** $\{A\}$
- **Conjunto de Estados $Q$:** $\{q_0, q_1, q_f\}$
- **Estado Inicial:** $q_0$
- **Conjunto de Estados Finais $F$:** $\{q_f\}$
- **Função de Transição $\delta$:**
  - $\delta(q_0, a, \epsilon) = \{(q_0, A)\}$: Empilha 'A' para cada 'a' lido.
  - $\delta(q_0, \epsilon, \epsilon) = \{(q_1, \epsilon)\}$: Transição não-determinística para começar a leitura de 'b's.
  - $\delta(q_1, b, A) = \{(q_1, \epsilon)\}$: Desempilha 'A' para cada 'b' lido.
  - $\delta(q_1, \epsilon, ?) = \{(q_f, \epsilon)\}$: Transição para o estado final se a pilha estiver vazia.

**Descrição do Processo**

1. **Leitura de 'a's:**
   - No estado $q_0$, para cada 'a' lido, um 'A' é empilhado. Isso continua até que não haja mais 'a's para ler.
   - A transição não-determinística permite mover para o estado $q_1$ sem consumir um símbolo de entrada, marcando o início da verificação de 'b's.

2. **Leitura de 'b's e Verificação da Pilha:**
   - No estado $q_1$, cada 'b' deve corresponder a um 'A' na pilha. Se a correspondência é correta, 'A' é desempilhado.
   - A transição para o estado final $q_f$ ocorre apenas se a pilha estiver vazia após todas as verificações, garantindo que o número de 'a's e 'b's é igual.

**Visualização da Função de Transição**

A função $\delta$ abrange todas as combinações possíveis de entradas e condições da pilha, demonstrando sua totalidade. Essas transições demonstram o não-determinismo, especialmente a transição de $q_0$ para $q_1$ que ocorre sem ler a entrada ou alterar a pilha. Isso permite flexibilidade no momento da mudança de leitura de 'a's para 'b's, dependendo unicamente da decisão interna do autômato, não de entradas externas ou condições da pilha.

1. **Empilhamento durante a Leitura de 'a's**:
   - **Transição:** $\delta(q_0, a, \epsilon) = \{(q_0, A)\}$
   - **Descrição:** No estado inicial $q_0$, a cada símbolo 'a' lido da entrada, um símbolo 'A' é empilhado na pilha. Isso continua até que todos os 'a's sejam lidos. Esta ação não depende de símbolos na pilha, demonstrando uma operação pura de empilhamento.

2. **Transição Não-determinística para Leitura de 'b's**:
   - **Transição:** $\delta(q_0, \epsilon, \epsilon) = \{(q_1, \epsilon)\}$
   - **Descrição:** Ainda no estado $q_0$, existe uma transição não-determinística que permite ao autômato mudar para o estado $q_1$ sem ler um símbolo da entrada ou modificar a pilha. Isso é essencial para começar a leitura dos 'b's após todos os 'a's terem sido lidos.

3. **Desempilhamento durante a Leitura de 'b's**:
   - **Transição:** $\delta(q_1, b, A) = \{(q_1, \epsilon)\}$
   - **Descrição:** No estado $q_1$, para cada 'b' lido, um símbolo 'A' é removido da pilha. Este desempilhamento continua para cada 'b' até que a pilha esteja vazia.

4. **Checagem de Pilha Vazia e Aceitação**:
   - **Transição:** $\delta(q_1, \epsilon, ?) = \{(q_f, \epsilon)\}$
   - **Descrição:** Após todos os 'b's serem lidos, se a pilha está vazia, o autômato move-se ao estado final $q_f$, aceitando a palavra se não restam mais símbolos a ler.

**Critérios de Parada em Autômatos com Pilha:**

Autômatos com Pilha (APs) têm várias maneiras de finalizar o processamento de uma entrada. Eles podem aceitar, rejeitar ou até mesmo entrar em um ciclo infinito, dependendo das condições abaixo:

1. **Aceitação:**
   - **Por estado final:** O AP pode parar se atingir um estado final e não houver mais entrada para processar. Isso significa que a palavra foi aceita pelo autômato.
   - **Por pilha vazia:** Alguns APs são configurados para aceitar quando a pilha é completamente consumida, independentemente do estado em que se encontram.

2. **Rejeição:**
   - Um AP rejeita uma entrada se todos os caminhos de computação possíveis levarem a configurações que não podem mais avançar (por exemplo, não há transições aplicáveis) ou se terminarem em estados não-finais sem consumir toda a entrada.

3. **Loop Infinito:**
   - Um AP pode entrar em um loop infinito se, em pelo menos um de seus caminhos de computação, ele repetir a mesma configuração indefinidamente sem consumir a entrada. Se qualquer caminho entra em loop infinito, considera-se que o AP pode potencialmente nunca parar para essa entrada.

**Exemplos de Comportamento:**
- **Aceitação por estado final:** Um AP que lê strings binárias e verifica se têm números iguais de 0s e 1s pode parar em um estado final se a condição for verdadeira.
- **Rejeição:** Um AP que verifica parênteses balanceados rejeitará a entrada `)(` pois não pode processá-la corretamente.
- **Loop Infinito:** Um AP que adiciona símbolos à pilha por cada 0 lido e nunca os remove pode entrar em loop infinito com a entrada `000...`.

### Exemplo 1

**Autômato com Pilha para $L = \{ a^n b^n | n \geq 0 \}$**:

A linguagem $L$ consiste em strings onde o número de 'a's é igual ao número de 'b's. Um Autômato com Pilha (AP) pode reconhecer essa linguagem usando uma pilha para contar os 'a's e verificar se eles são seguidos pelo mesmo número de 'b's.

**Descrição do Autômato:**
- **Estados:**
  - $q_0$: Estado inicial onde o AP começa a operação e empilha cada 'a' lido.
  - $q_1$: Estado onde o AP começa a desempilhar por cada 'b' lido.
  - $q_f$: Estado final que indica que a palavra foi aceita.
- **Transições:**
  - $ (a, \epsilon, B) $: Em $q_0$, para cada 'a' lido, empilha um 'B'.
  - $ (b, B, \epsilon) $: Em $q_1$, para cada 'b' lido, desempilha um 'B'.
  - $ (?, ?, \epsilon) $: Transições que permitem mover-se entre $q_0$ e $q_1$ sem modificar a pilha, e de $q_1$ para $q_f$ se a pilha estiver vazia.
- **Alfabeto de Pilha:**
  - Utiliza o símbolo 'B' como marcador na pilha.

**Funcionamento:**
- O autômato inicia em $q_0$ e empilha um 'B' para cada 'a' lido, garantindo que cada 'a' tenha um correspondente 'b' posterior.
- Ao ler o primeiro 'b', o autômato transita para $q_1$ e começa a desempilhar, verificando se o número de 'b's corresponde ao número de 'a's empilhados.
- Se todos os 'b's forem lidos e a pilha estiver vazia, o autômato transita para $q_f$, indicando que a palavra pertence à linguagem $L$.



### Exemplo 3 - Autômato com Pilha para a Linguagem $L = \{ww^r | w \in \{a, b\}^*\}$

Esta linguagem consiste em palavras seguidas por suas reversas, como 'abba'. Um autômato com pilha pode reconhecer essa linguagem usando a pilha para armazenar a primeira metade da palavra e depois verificar se a segunda metade é a reversa da primeira.

**Descrição do Autômato:**

- **Estados**: 
  - $q_0$: Estado inicial onde os símbolos são lidos e empilhados.
  - $q_1$: Estado de transição para começar a comparar os símbolos com o topo da pilha.
  - $q_f$: Estado final que indica que a palavra é aceita.

- **Transições**:
  - $(q_0, a, \epsilon) \to (q_0, a)$: Lê 'a', empilha 'a'.
  - $(q_0, b, \epsilon) \to (q_0, b)$: Lê 'b', empilha 'b'.
  - $(q_0, \epsilon, \epsilon) \to (q_1, \epsilon)$: Transição sem consumo para começar a desempilhar.
  - $(q_1, a, a) \to (q_1, \epsilon)$: Lê 'a', desempilha 'a'.
  - $(q_1, b, b) \to (q_1, \epsilon)$: Lê 'b', desempilha 'b'.
  - $(q_1, ?, \epsilon) \to (q_f, \epsilon)$: Se a pilha estiver vazia e não houver mais entrada, aceita a palavra.

## Slide 3 - Hierarquia de Chomsky

A Hierarquia de Chomsky categoriza as linguagens formais em quatro níveis de complexidade:

1. **Linguagens Regulares**
   - São as linguagens mais simples na hierarquia.
   - Podem ser reconhecidas por autômatos finitos e expressadas por expressões regulares.
   - Exemplo: Linguagens que aceitam sequências de dígitos ou identificadores em linguagens de programação.

2. **Linguagens Livres de Contexto**
   - Um nível acima das regulares, podem ser reconhecidas por autômatos com pilha.
   - Usadas para definir a sintaxe de linguagens de programação através de gramáticas livres de contexto.
   - Exemplo: A linguagem que inclui todas as cadeias corretamente balanceadas de parênteses.

3. **Linguagens Sensíveis ao Contexto**
   - Podem ser reconhecidas por máquinas de Turing com fita limitada.
   - São mais poderosas que as livres de contexto e podem definir linguagens que dependem de um contexto específico para serem validadas.
   - Exemplo: A linguagem que consiste em strings com igual número de 'a's, 'b's e 'c's em sequência ($a^n b^n c^n$).

4. **Linguagens Recursivamente Enumeráveis**
   - No topo da hierarquia, podem ser reconhecidas por Máquinas de Turing sem restrição de espaço na fita.
   - Incluem todas as linguagens que uma máquina de Turing é capaz de reconhecer.
   - Exemplo: A linguagem do problema da parada.

### O Problema da Parada 

O **Problema da Parada** é um conceito fundamental na teoria da computação, 
que foi introduzido pelo matemático Alan Turing em 1936. Este problema nos ajuda a entender os limites do que os computadores podem e não podem fazer.
Esse problema questiona se é possível criar um algoritmo que, ao receber como entrada qualquer descrição de um programa e um input inicial para esse programa, é capaz de determinar corretamente se o programa para ou continua a executar indefinidamente (entra em loop) quando executado com esse input.

**O que é o Problema da Parada?**

- Imagine que você tenha um programa de computador e você quer saber se esse programa vai parar de rodar (terminar) ou se vai continuar rodando para sempre
- O "Problema da Parada" pergunta se existe um outro programa, que podemos chamar de "juiz", que pode analisar qualquer programa de computador e dizer com certeza se ele vai parar ou não.

**Indecidibilidade do Problema** 

- Turing provou que não existe uma solução geral para o problema da parada. Isso significa que não há um algoritmo que possa determinar, para todas as possíveis máquinas de Turing e entradas, se a máquina irá parar. 
- A prova envolve uma técnica de argumentação chamada **diagonalização**, que basicamente constrói um cenário de auto-referência onde uma hipotética máquina de Turing que decide o problema da parada é levada a uma contradição.

**Por que é importante?**

A resposta que Turing encontrou foi surpreendente: ele provou que **não é possível** construir tal programa "juiz". Isso significa que existem limites para o que podemos prever com algoritmos. Este resultado tem implicações profundas, mostrando que existem problemas que são intrinsecavelmente complexos ou indecidíveis — não podemos resolver tudo com computadores!

**Como isso afeta a vida real?**

A indecidibilidade do problema da parada tem várias consequências práticas:

- **Desenvolvimento de Software:** Não podemos criar um programa que verifique automaticamente e com 100% de certeza se outro programa tem bugs ou se vai terminar.
- **Segurança:** Não é possível garantir automaticamente que um programa é seguro e não será malicioso.
- **Inovação Tecnológica:** Este problema nos lembra que sempre haverá desafios que precisam de novas teorias e abordagens na ciência da computação.

**Conclusão**

O Problema da Parada é mais do que uma curiosidade matemática; é uma lembrança de que, mesmo em uma era dominada pela tecnologia, ainda existem mistérios que desafiam nossa capacidade de entender e controlar completamente o mundo digital.

### Explicação da Prova de Turing

A diagonalização é um método matemático usado para demonstrar a existência de limites no que pode ser alcançado com procedimentos computacionais. Turing aplicou este conceito para mostrar que existem problemas que as máquinas de Turing, e por extensão qualquer computador moderno, não podem resolver.

**Passos da Prova:**
1. **Suposição Inicial:**
   Suponha a existência de uma máquina de Turing, chamada "Decididor", que pode decidir o problema da parada para qualquer máquina de Turing. Ou seja, para qualquer máquina de Turing $M$ e entrada $w$, o Decididor pode determinar corretamente se $M$ irá parar ou continuar executando indefinidamente.

2. **Construção de uma Máquina de Turing Especial (D):**
   Construímos uma máquina de Turing $D$ que usa o Decididor para determinar o comportamento de outras máquinas, incluindo a si mesma.

3. **Auto-Referência:**
   $D$ opera da seguinte forma: 
   - recebe como entrada a descrição de qualquer máquina de Turing $M$, e consulta o Decididor para saber se $M$ executada com sua própria descrição irá parar. 
   - se o Decididor prevê que $M$ irá parar, $D$ entra em um loop infinito. 
   - se prevê que $M$ não irá parar, então $D$ para imediatamente.

4. **Contradição:**
   O que ocorre quando $D$ é executada com sua própria descrição como entrada? De acordo com a lógica da construção de $D$:
   - Se o Decididor diz que $D$ para, então $D$ deve entrar em loop (contradição, pois o Decididor deveria estar correto).
   - Se o Decididor diz que $D$ entra em loop, então $D$ para (outra contradição pelo mesmo motivo).


5. **Conclusão:**
   Essa contradição leva à conclusão de que a suposição inicial de tal Decididor pode existir é falsa. Portanto, concluímos que não existe tal máquina de Turing que decide o problema da parada para todas as possíveis máquinas de Turing e entradas, tornando o problema da parada indecidível.

**Significado da Prova**

Esta prova não apenas destaca a indecidibilidade do problema da parada mas também ilustra um princípio mais amplo sobre os limites da computação, ajudando a entender que existem problemas insolúveis por algoritmos, um conceito fundamental na teoria da computação.


#### Explicação Simplificada do Problema da Parada



Imagine que temos uma máquina mágica, chamada Decididor, que supostamente pode responder se qualquer programa de computador vai parar de rodar ou se vai continuar rodando para sempre, dada qualquer entrada específica para esse programa.

Agora, criamos um programa especial chamado $D$. Este programa é projetado para fazer exatamente o oposto do que o Decididor diz. Veja como ele funciona:

1. **Entrada:** O programa $D$ recebe o nome de um programa qualquer, vamos chamar de $M$.

2. **Consulta ao Decididor:** $D$ pergunta ao Decididor: "Se eu rodar o programa $M$ com o nome dele mesmo como entrada, ele vai parar?"

3. **Ação baseada na resposta:**
   - **Se o Decididor diz que $M$ vai parar,** então $D$ decide entrar em um loop infinito.
   - **Se o Decididor diz que $M$ não vai parar,** então $D$ decide parar imediatamente.

4. **Testando o Decididor com $D$:** A parte interessante acontece quando usamos $D$ como entrada para si mesmo. Ou seja, perguntamos ao Decididor: "O programa $D$, rodando com a entrada '$D$', vai parar?"

   - **Se o Decididor diz que $D$ vai parar,** $D$ entra em um loop infinito (o oposto do que o Decididor disse).
   - **Se o Decididor diz que $D$ não vai parar,** então $D$ para imediatamente (novamente o oposto do que foi previsto).

**Conclusão**

Essa contradição mostra que o Decididor não pode existir. Se ele existisse e pudesse sempre prever corretamente se um programa vai parar ou não, não poderia errar. Mas como $D$ sempre faz o oposto do que o Decididor prevê, isso prova que não podemos ter uma máquina ou programa que decida corretamente o problema da parada para todos os programas e entradas possíveis. Essa é a natureza "indecidível" do problema da parada.

## Slide 6 - Máquina de Turing com Fita Infinita

Em 1936, Alan Turing introduziu o conceito de uma Máquina de Turing, que é amplamente aceito como um modelo para algoritmos ou funções computáveis. Este modelo ajudou a fundamentar a ciência da computação teórica.

Para ser representado por uma Máquina de Turing, um algoritmo deve satisfazer algumas propriedades críticas:

- A **descrição do algoritmo** deve ser finita.
- Os **passos do algoritmo** devem ser:
  - Discretos
    - Os passos de um algoritmo precisam ser claramente definidos e distintos entre si. Cada operação deve ser uma ação individual e bem definida, que não se sobrepõe a outras ações. Isso garante que o algoritmo seja sequencial e cada passo seja compreensível e isolado dos demais.  

  - Executáveis mecanicamente
    - Um passo é considerado executável mecanicamente se puder ser realizado por uma máquina ou um processo automatizado sem intervenção humana. Isso implica que cada ação do algoritmo deve ser algo que uma máquina, como uma Máquina de Turing, possa realizar. Esta propriedade assegura que o algoritmo pode ser implementado e executado em um sistema computacional.  

  - De tempo finito
    - Cada passo do algoritmo deve ser completado em um tempo finito, ou seja, deve haver uma garantia de que cada operação se conclua em uma quantidade de tempo limitada e previsível. Isso é crucial para a viabilidade de um algoritmo, assegurando que ele não entre em loop infinito durante uma única operação ou passo.

A Máquina de Turing é composta por três componentes principais:

1. **Fita**:
   - Serve como dispositivo de entrada, saída e memória de trabalho.
   - É conceitualmente infinita à direita, proporcionando um espaço ilimitado para operações.

2. **Unidade de Controle**:
   - Representa o estado atual da máquina.
   - Possui uma cabeça de leitura/gravação que pode acessar uma célula da fita por vez.
   - A cabeça pode se mover para a esquerda ou para a direita, permitindo operações sobre a fita.

3. **Programa ou Função de Transição**:
   - Define as ações de leitura e escrita na fita.
   - Determina o movimento da cabeça de leitura/gravação e as transições de estado da máquina.

Este modelo abstrato permite a simulação de qualquer algoritmo que possa ser descrito de maneira finita, tornando a Máquina de Turing um poderoso instrumento teórico para explorar os limites da computação.

## Slide 8 - Máquina de Turing com Fita Infinita - descrição formal

Uma Máquina de Turing com Fita Infinita **M** é definida por uma 8-tupla:

$$
M = \{ \Sigma, Q, \delta, q_0, F, V, \beta, \circledast \}
$$

onde cada componente tem o seguinte significado:

- $\Sigma$: **Alfabeto de símbolos de entrada**. Contém os símbolos que a máquina pode ler.
- $Q$: **Conjunto finito de estados** do autômato.
- $q_0$: **Estado inicial** da máquina.
- $F$: **Conjunto de estados finais**. Se a máquina entra em um desses estados, ela para.
- $V$: **Alfabeto auxiliar**, usado para símbolos adicionais na fita que não são parte do alfabeto de entrada.
- $\beta$: **Símbolo especial branco**, usado para indicar espaços vazios na fita.
- $\circledast$: **Marcador de início da fita**, ajuda a indicar o começo da fita para a máquina.
- $\delta$: **Função de transição**, definida como:
  $$
  \delta: Q \times (\Sigma \cup V \cup \{ \beta, \circledast \}) \to Q \times (\Sigma \cup V \cup \{ \beta, \circledast \}) \times \{ E, D \}
  $$
  Esta função determina o próximo estado e ação da máquina com base no estado atual e no símbolo lido.  
  
**Componentes da Função de Transição:**
- **$Q$**: Conjunto de todos os estados possíveis na máquina, com quantidade finita.
- **$\Sigma \cup V \cup \{ \beta, \circledast \}$**: 
  - indica, como resultado da função de transição, a máquina pode mover-se para qualquer estado dentro de $Q$ e pode escrever qualquer símbolo de $\Sigma$, $V$, $\beta$, ou $\circledast$ na posição atual da fita.
  - o conjunto de possíveis caracteres que a máquina pode ler. Inclui:
    - **$\Sigma$**: alfabeto de entrada, contendo os símbolos que a máquina foi projetada para ler.
    - **$V$**: alfabeto auxiliar, que pode conter símbolos adicionais usados pela máquina mas que não fazem parte da entrada inicial.
    - **$\beta$**: é um símbolo especial usado para indicar um espaço vazio na fita, também conhecido como o "símbolo branco".
    - **$\circledast$**: é o símbolo que marca o início da fita, ajudando a máquina a reconhecer o começo de sua área de trabalho.
- **$\{ E, D \}$**: Indica as direções em que a cabeça de leitura/escrita da máquina pode mover-se após a execução da transição.
  - **$E$** (Esquerda): Move a cabeça de leitura/gravação um espaço para a esquerda.
  - **$D$** (Direita): Move a cabeça de leitura/gravação um espaço para a direita.  
- $\to$: Indica que a função de transição mapeia cada par de estado e símbolo lido para um novo conjunto de instruções.  

Essencialmente, a função de transição diz/pergunta "dado o estado atual e o símbolo que estou lendo, para qual novo estado devo ir, que símbolo devo escrever no lugar do atual, e em qual direção devo mover a cabeça de leitura/escrita?".

**Funcionamento**

- A máquina começa no estado inicial $q_0$ e lê o símbolo na posição atual da fita.
- A função $\delta$ é consultada para determinar a ação: mover a cabeça de leitura/gravação para a esquerda (**E**) ou direita (**D**), escrever um símbolo na fita, e mudar para um novo estado.
- Se a máquina alcança um estado em $F$, ela para, indicando que a entrada foi aceita ou rejeitada com base no estado final alcançado.

Graficamente, a função programa pode ser representada como:

**Condições de Parada:**

- **Estado Final:** A máquina para e a palavra é aceita se, durante a computação, a máquina alcança um estado que está definido como final no conjunto $F$.
- **Função Programa Indefinida:** Se a função de transição $\delta$ não possui uma definição para a combinação atual do estado e o símbolo lido, a máquina para e a palavra é rejeitada.
- **Movimento Ilegal:** Se a máquina tenta mover a cabeça de leitura/escrita para a esquerda quando já está na célula mais à esquerda da fita, a máquina também para e a palavra é rejeitada.

**Funcionamento Além da Parada:**

- **Loop Infinito:** Se nenhuma das condições de parada for satisfeita e a máquina continua operando indefinidamente sem atingir um estado final, a máquina está em um loop infinito. Isso pode ocorrer devido a ciclos na função de transição onde a máquina continua repetindo um conjunto de estados e operações sem progresso.

**Relevância na Computabilidade:**

- A habilidade de uma Máquina de Turing de continuar operando indefinidamente ou parar com base na entrada é central para questões de decidibilidade em teoria da computação. Problemas para os quais não existe um algoritmo de Turing que sempre para são chamados de indecidíveis.

**1- Estado $q_0$ e Transições**

O estado $q_0$ é o estado inicial da Máquina de Turing. Neste estado, a máquina realiza a preparação inicial da fita e começa o processo de transformação dos símbolos de entrada. As transições para este estado são detalhadas como segue:

  - Lê 'B', mantém 'B', move para direita → continua em $q_3$

**Processo Geral da Máquina de Turing para $\{a^n b^n | n \geq 0\}$**

1. Inicialização:
- A máquina inicia no estado inicial $q_0$, com a cabeça de leitura na posição do primeiro caractere da fita.

2. Marcação Alternada de 'a's e 'b's:
- Para cada 'a' lido, a máquina:
  - Substitui o primeiro 'a' (da esquerda para a direita) por 'A' para marcar que foi processado.
  - Move-se para a direita até encontrar o primeiro 'b' não marcado.
  - Ao encontrar um 'b':
    - Substitui o 'b' por 'B', indicando que foi pareado com um 'A'.
    - Move-se para a esquerda para retornar ao próximo primeiro 'a' não marcado.

**Validação e Retorno**
- A máquina retorna para a esquerda até encontrar o marcador de início ou o próximo primeiro 'a' não marcado.
- Se um 'A' é encontrado durante o retorno, a máquina retorna ao estado inicial e continua movendo-se para a direita até encontrar um 'a' não processado, começando o ciclo novamente de **Marcação Alternada de 'a's e 'b's**.

**Finalização:**
- No estado $q_3$, a máquina verifica se todos os 'B's foram pareados movendo-se para a direita. Se um espaço em branco (fim da fita) é encontrado após todos os 'B's, a máquina transita para o estado de aceitação $q_4$, confirmando que a entrada está correta com números iguais de 'a's e 'b's.
- Se discrepâncias são encontradas ou sobram 'b's sem pareamento, a entrada é rejeitada.

## Slide 12 - Hipótese de Church ou Hipótese de Turing-Church

A **Hipótese de Church-Turing** propõe uma definição fundamental sobre o conceito de computabilidade:
- **Enunciado**: A capacidade de computação representada pela Máquina de Turing encapsula todas as possibilidades de cálculo que podem ser realizadas por qualquer dispositivo computacional. 
    - Isso sugere que qualquer função que possa ser calculada por algum algoritmo pode ser calculada por uma Máquina de Turing.
    - Esta hipótese, implica que a Máquina de Turing é capaz de simular qualquer outro modelo de computação concebível, desde que este modelo funcione dentro dos limites dos procedimentos algorítmicos.

A **Hipótese de Turing** é amplamente aceita porque nenhum modelo computacional conhecido até agora demonstrou um poder computacional superior ao da Máquina de Turing. Vários modelos foram propostos, cada um tentando estender ou variar as capacidades originais da Máquina de Turing, mas todos esses modelos têm se mostrado equivalentes em poder computacional. Estes incluem:

- **Autômato com Múltiplas Pilhas**: Variantes de autômatos com pilha única, equipados com mais de uma pilha para armazenamento temporário.
- **Máquina de Turing Não-determinística**: Uma abstração que pode seguir muitos caminhos de computação simultaneamente.
- **Máquina de Turing com Fita Infinita à Esquerda e à Direita**: Permite movimentação e operações em ambas as direções indefinidamente.
- **Máquina de Turing com Múltiplas Fitas**: Utiliza múltiplas fitas, cada uma com sua própria cabeça de leitura/gravação, para realizar operações mais complexas.
- **Máquina de Turing Multidimensional**: Operações não são restritas a uma fita linear, podendo explorar arranjos multidimensionais de células.
- **Máquina de Turing com Múltiplas Cabeças**: Uma única fita é acessada por múltiplas cabeças de leitura/gravação, proporcionando mais eficiência em certas operações.
- **Combinações de Modificações sobre a Máquina de Turing**: Inclui qualquer número de variações e combinações das modificações mencionadas acima.

Esses modelos, apesar de suas diferenças estruturais ou operacionais, não conseguem resolver problemas que uma Máquina de Turing padrão não possa resolver, reforçando a centralidade e relevância da hipótese de Turing na ciência da computação teórica.

## Slide 14 - Definição de Linguagem Recursivamente Enumerável (Tipo 0)

**Definição**: Uma linguagem que é aceita por uma Máquina de Turing é conhecida como uma **Linguagem Recursivamente Enumerável** ou do **Tipo 0**. 
  - O termo "enumerável" sugere que qualquer palavra na linguagem pode ser gerada ou "listada" por uma Máquina de Turing. A geração dessas palavras ocorre de maneira recursiva, onde a máquina pode repetir um conjunto definido de instruções para produzir todas as palavras possíveis na linguagem.
  - ou seja, uma linguagem é recursivamente enumerável se existe uma Máquina de Turing que pode aceitar todas as cadeias na linguagem. 

#### Exemplos de Linguagens Recursivamente Enumeráveis

- **Linguagem $\{ a^n b^n | n \geq 0 \}$**
  - Inclui sequências como '', 'ab', 'aabb', 'aaabbb', etc., onde o número de 'a's é igual ao número de 'b's.
  - Máquina de Turing Específica:
    - Para esta linguagem, uma Máquina de Turing pode ser construída para ler sequencialmente 'a's e 'b's, garantindo que o número de 'a's seja igual ao de 'b's. A máquina pode, por exemplo, substituir cada 'a' lido por um símbolo especial (como 'A'), avançar até os 'b's correspondentes, substituir cada 'b' por outro símbolo especial (como 'B'), e então retornar para verificar se há mais 'a's. Se ao final todos os 'a's e 'b's forem substituídos e validados de forma equilibrada, a entrada é aceita.
  - Enumerabilidade: 
    - Uma Máquina de Turing pode gerar todas as strings possíveis desta linguagem ao começar com zero ou mais 'a's, seguido por um número igual de 'b's, e verificar cada string para aceitação.

- **Linguagem $\{ w | w$ tem o mesmo número de símbolos 'a' e 'b' $\}$**
  - Abrange palavras com igual quantidade de 'a's e 'b's, independentemente da ordem.
  - Máquina de Turing Complexa: 
    - Uma máquina para esta linguagem precisaria contar e comparar a quantidade de 'a's e 'b's, o que pode ser feito usando uma série de estados ou fitas auxiliares para armazenar contagens temporárias. Cada configuração da máquina que termina com contagens iguais para 'a' e 'b' indica uma string aceita.
  - Enumerabilidade: 
    - Uma Máquina de Turing pode listar todas as strings possíveis de 'a's e 'b's em várias ordens e tamanhos, verificando cada uma para garantir que a quantidade de 'a's seja igual à de 'b's antes de aceitar a string.

- **Linguagem $\{ a^i b^j c^k | i = j$ ou $j = k \}$**
  - Composta por strings como 'abc', 'aabbc', 'abcc', 'abbcc', etc., onde ou o número de 'a's é igual ao de 'b's, ou o de 'b's é igual ao de 'c's.
  - Máquina de Turing com Condições: 
    - Esta linguagem requer que a Máquina de Turing verifique duas condições: se os números de 'a's e 'b's são iguais, ou se os números de 'b's e 'c's são iguais. Isso pode ser alcançado usando múltiplas fitas ou múltiplos cabeçotes de leitura para rastrear e comparar esses contadores.
    - Enumerabilidade: A máquina pode sistematicamente construir e testar strings para essas condições, enumerando todas as combinações possíveis de 'a's, 'b's e 'c's e verificando cada uma conforme as regras definidas.

**Por que são Recursivamente Enumeráveis?**
Porque existe uma Máquina de Turing que, ao operar sobre uma entrada infinita de símbolos, pode eventualmente produzir (enumerar) qualquer string válida dentro da linguagem se for dada tempo suficiente. Em outras palavras, para qualquer string que pertença à linguagem, há um procedimento de Máquina de Turing que a reconhecerá eventualmente.


## Slide 15 - Linguagem Recursivamente Enumerável e Looping

- Existem **Linguagens Recursivamente Enumeráveis** para as quais algumas palavras podem entrar em **loop** quando processadas por uma Máquina de Turing.
- As linguagens desta classe que garantem uma resposta (aceitação ou rejeição) para todas as entradas são chamadas de **Linguagens Recursivas**. Uma linguagem é recursiva se uma Máquina de Turing:
    - **ACEITA(M) = L**: A máquina $M$ aceita todas as palavras em $L$.
    - **REJEITA(M) = $\Sigma^* - L$**: A máquina $M$ rejeita todas as palavras que não estão em $L$.  
    
**Exemplos**
1. $\{a^n b^n | n \geq 0\}$:
   - Linguagem das palavras formadas por 'n' ocorrências de 'a' seguidas por 'n' ocorrências de 'b'.
   - Exemplo de palavras: '', 'ab', 'aabb', 'aaabbb', etc.
   - A máquina terminará sua execução após verificar todos os caracteres, sem entrar em loop.

2. $\{a^n b^n c^n | n \geq 0\}$:
   - Linguagem das palavras formadas por 'n' ocorrências de 'a', seguidas por 'n' ocorrências de 'b', seguidas por 'n' ocorrências de 'c'.
   - Exemplo de palavras: '', 'abc', 'aabbcc', 'aaabbbccc', etc.
   - Similarmente à linguagem anterior, uma Máquina de Turing para essa linguagem começaria verificando a sequência de 'a's, marcando o primeiro à medida que avança. Após, avançaria e começaria a verificar e marcar o primeiro 'b', e em seguida, o primeiro 'c', garantindo que o número de cada símbolo seja igual e que estejam na ordem correta. Sem seguida, retornaria ao início e repetiria o processo. Qualquer desvio da estrutura esperada resulta na rejeição da palavra. A máquina também termina depois de processar toda a entrada, garantindo que todas as operações sejam finitas e decidíveis.

3. $\{ w | w \in \{a, b\}^*$ e tem o dobro de símbolos a que b $\}$:
   - Linguagem das palavras sobre o alfabeto $\{a, b\}$ onde o número de 'a's é exatamente o dobro do número de 'b's.
   - Exemplo de palavras: 'aab', 'aaaabb', 'aaaaaabbb', etc.
   - Como seria uma Máquina de Turing para reconhecer essa linguagem?

**Exemplo de Máquina de Turing que Entra em Loop**

Consideremos uma Máquina de Turing simplificada com as seguintes características:

- **Estados**: $\{q_0, q_1\}$
- **Alfabeto de fita**: $\{0, 1\}$
- **Função de transição**:
  - $(q_0, 1) \to (q_0, 1, D)$: Lê '1', escreve '1', move para direita, permanece em $q_0$.
  - $(q_0, 0) \to (q_1, 0, D)$: Lê '0', escreve '0', move para direita, transita para $q_1$.
  - $(q_1, 0) \to (q_0, 0, D)$: Lê '0', escreve '0', move para direita, retorna para $q_0$.
  - $(q_1, 1) \to (q_1, 1, D)$: Lê '1', escreve '1', move para direita, permanece em $q_1$.

**Situação de Loop**
Suponhamos que a fita inicial seja `1000...` (um '1' seguido por um número infinito de '0's). A máquina entrará em um ciclo infinito alternando entre os estados $q_0$ e $q_1$ enquanto lê e escreve '0' sem parar. Aqui está o que acontece:

1. **Início em $q_0$**:
   - **Fita**: `[1]0000...` (cabeça de leitura sobre o primeiro '1')
   - **Fita**: `1[0]000...` (cabeça de leitura sobre o primeiro '0')
2. **Transição para $q_1$**:
   - **Fita**: `10[0]00...` (move para a direita)
3. **Transição para $q_0$**:
   - **Fita**: `100[0]0...` (move para a direita)
4. **Repete as etapas 2 e 3**:
   - Continua movendo para a direita, alternando estados, sem alterar os símbolos da fita ou finalizar.

Este ciclo nunca leva a máquina a um estado de parada ou a um estado onde uma ação diferente ocorra, resultando em um loop infinito. Este exemplo ilustra como uma simples Máquina de Turing pode falhar em decidir ou parar, caracterizando a linguagem como recursivamente enumerável, mas não recursiva, já que não há garantia de parada para todas as entradas.

## Slide 16 - Gramática Irrestrita

**Como teorema, tem $L$ é um Linguagem Recursivamente Enumerável se, e somente se, $L$ é gerada por uma Gramática Irrestrita.**
- Podem modelar qualquer linguagem que seja computável por uma Máquina de Turing, incluindo linguagens que não podem ser geradas por qualquer outro tipo de gramática ou autômato.
- É aquela em que não há restrições sobre o formato de suas regras. 
   - não tem restrições quanto à forma de suas produções
   - exceto que o lado esquerdo deve conter pelo menos um símbolo. 
   - as produções podem ter qualquer forma $u \to v$, onde:
      - $u$ é uma cadeia não vazia de terminais e não terminais
      - $v$ é qualquer cadeia de terminais e não terminais. 
   - em outras palavras, $u$ pode ser maior, igual ou menor que $v$, permitindo que a gramática descreva operações que encolhem, mantêm ou expandem a cadeia de entrada.
   - uma regra de produção pode ter qualquer forma, como $u \to v$, onde $u$ é uma sequência de terminais e não terminais e $v$ é outra sequência qualquer. Essas regras podem ser aplicadas independentemente dos símbolos que cercam o segmento $u$ na string que está sendo reescrita.
   - Por exemplo, se tivermos a regra $AB \to CD$, podemos aplicar essa regra independentemente do que vem antes de $A$ ou depois de $B$ na string.
- Exemplos de Aplicação da Regra
$$
AB \to CD
$$
   - **String Inicial**: `XYZABW`
   - **Aplicação da Regra**: A parte `AB` é substituída por `CD`.
   - **Resultado**: `XYZCDW`



### Exemplo de Gramática Irrestrita que gera a Linguagem: $\{a^n b^n c^n | n \geq 1\}$

**Variáveis**: ${S, A, B, C, X}$
**Alfabeto**: $\{a, b, c\}$
**Regras de Produção**:
- $S \rightarrow XABC \mid \epsilon$
- $Xa \rightarrow aX$
- $Xb \rightarrow bX$
- $Xc \rightarrow cX$
- $XA \rightarrow aA$
- $AB \rightarrow bB$
- $BC \rightarrow cC$
- $XC \rightarrow c$
- $A \rightarrow a$
- $B \rightarrow b$
- $C \rightarrow c$

**Explicação das Regras**
- A regra $S \rightarrow XABC \mid \epsilon$ introduz um marcador $X$ que ajuda na construção da palavra final.
- As regras como $Xa \rightarrow aX$, $Xb \rightarrow bX$, e $Xc \rightarrow cX$ permitem mover o marcador $X$ para a direita, preparando o terreno para a inserção dos símbolos apropriados em cada passo.
- As regras $XA \rightarrow aA$, $AB \rightarrow bB$, e $BC \rightarrow cC$ são usadas para inserir múltiplas ocorrências de 'a', 'b', e 'c' respectivamente, aumentando cada segmento da palavra conforme necessário.
- Finalmente, regras como $XC \rightarrow c$, $A \rightarrow a$, $B \rightarrow b$, e $C \rightarrow c$ terminam a construção de cada bloco de letras, eliminando os marcadores intermediários e substituindo-os pelos símbolos finais.

**Por Que Não é Sensível ao Contexto**
- A gramática permite regras que manipulam e alteram a cadeia de entrada de maneiras que não são possíveis em gramáticas sensíveis ao contexto, que exigem que o contexto ao redor de um símbolo seja preservado a menos que sejam feitas substituições simétricas. Aqui, símbolos podem ser adicionados e o marcador movido independentemente dos elementos adjacentes.

## Slide 18 - Gramáticas Sensíveis ao Contexto ou do Tipo 1

Uma **Gramática Sensível ao Contexto (GSC)** $G$ é definida pela quádrupla:
$$G = (V, T, P, S)$$
onde:
- $V$ é o conjunto de variáveis (não-terminais),
- $T$ é o conjunto de símbolos terminais,
- $P$ é o conjunto de regras de produção da forma $\alpha \to \beta$, onde
   - $\alpha$ é uma palavra de $(V \cup T)^+$, ou seja, o conjunto de todas as possíveis concatenações de um ou mais símbolos do conjunto  $V \cup T$, excetuando-se a palavra vazia. A notação $+$ indica que a palavra deve ter pelo menos um símbolo
   - $\beta$ é uma palavra de $(V \cup T)^*$, ou seja, denota o conjunto de todas as palavras (incluindo a palavra vazia) formadas a partir dos símbolos em $V \cup T$. A notação $*$ indica que a palavra pode ter zero ou mais símbolos
   - $|\alpha| \leq |\beta|$, excetuando-se, eventualmnte, para $S \rightarrow \epsilon$


Linguagens sensíveis ao contexto, também conhecidas como linguagens do tipo 1, podem ser compreendidas através de dois formalismos principais que destacam tanto sua geração quanto reconhecimento:

1. **Operacional ou Reconhecedor**:
   - **Representação**: Máquinas de Turing com Fita Limitada.
   - **Descrição**: Neste modelo, a computação é realizada por uma Máquina de Turing que opera dentro de limites fixos na fita, o que simula as restrições contextuais de produção nas gramáticas sensíveis ao contexto.

      - **Linguagem**: $\{a^i b^j c^k | i = j = k\}$
      - **Descrição**: Uma Máquina de Turing projetada para aceitar esta linguagem funcionaria com um segmento limitado da fita para garantir que o número de 'a's, 'b's e 'c's seja igual. A máquina marca cada 'a' lido e verificado com 'X', cada 'b' com 'Y', e cada 'c' com 'Z', e faz várias passagens para conferir a correspondência entre os símbolos.

2. **Axiomático ou Gerador**:
   - **Representação**: Gramáticas Sensíveis ao Contexto.
   - **Descrição**: Estas gramáticas permitem regras de produção onde a substituição de um símbolo pode depender de seu contexto (símbolos à esquerda e à direita). Isso difere das gramáticas livres de contexto, onde as regras de produção são aplicadas independentemente do contexto dos símbolos.  

      

### Exemplos de Gramática Sensível ao Contexto - Gramática que gera a Linguagem: $\{a^n b^n c^n | n \geq 1\}$

Um exemplo clássico de linguagem sensível ao contexto é a linguagem $\{a^n b^n c^n | n \geq 1\}$, que requer uma correspondência tripla de 'a's, 'b's, e 'c's em ordem sequencial.
   - A classificação das linguagens em termos da Hierarquia de Chomsky não é exclusiva, ou seja, uma linguagem pode pertencer a mais de uma classe ao mesmo tempo.
   - Essa linguagem é sensível ao contexto porque você precisa de um número igual de 'a's, 'b's e 'c's em uma ordem específica, o que implica uma restrição contextual na forma como os símbolos são organizados.
   - Ela é recursivamente enumerável porque pode ser gerada por uma máquina de Turing (ou uma gramática irrestrita)

$$G_2=(\{S, A, B, C\}, \{a, b, c\}, P, S)$$  
**Identificando as variáveis e Produções**
- **Variáveis**: $\{S, A, B, C\}$
- **Alfabeto**: $\{a, b, c\}$
- **Produções P**: 
   - $S \rightarrow aSBc ∣ abC$
   - $aB \rightarrow ab$
   - $bB \rightarrow bb$
   - $cB \rightarrow cc$
   - $bC \rightarrow bb$
   - $cC \rightarrow cc$
- **Descrição**: Esta gramática gera palavras começando com uma produção inicial $S$ que é substituída por $aSBC$ repetidamente, inserindo gradualmente 'a's, 'b's e 'c's na ordem correta. Quando o número desejado de repetições é atingido, a variável $S$ é substituída por $abc's$.
- A gramática descrita para a linguagem ${a^n b^n c^n | n \geq 1}$ é uma Gramática Sensível ao Contexto. Esse tipo de gramática é classificado como Tipo 1 na Hierarquia de Chomsky, que permite regras de produção na forma $\alpha A \beta \to \alpha \gamma \beta$, onde $A$ é uma variável e $\alpha$, $\beta$, e $\gamma$ são cadeias de variáveis e/ou terminais.
A palavra $aaabbbccc$ na Gramática 2 pode ser derivada da seguinte forma:

1. **Inicialização**:
   - Começamos com a variável inicial $S$.
   - **Derivação inicial**: 
     ```plaintext
     S
     ```
2. **Primeira Expansão de $S$**:
   - Aplicamos a regra $S \to aSBC$, adicionando uma camada de $a$, $B$, e $C$ enquanto mantemos uma cópia de $S$ para expansão futura.
   - **Derivação**:
     ```plaintext
     [aSBc]
     ```

3. **Segunda Expansão de $S$**:
   - Novamente, aplicamos $S \to aSBC$ na segunda cópia de $S$.
   - **Derivação**:
     ```plaintext
     a[aSBc]Bc
     ```

5. **Finalização de $S$**:
   - Substituímos a última cópia de $S$ por $abc$ usando $S \to abc$.
   - **Derivação**:
     ```plaintext
     aa[abC]BcBc
     ```

5. **Finalização de $S$**:
   - Aplicamos as regras de contexto para transformar $B$ e $C$:
      - $bC \rightarrow bb$
      - $bB \rightarrow bb$
      - $cB \rightarrow cc$
   - **Derivação**:
     ```plaintext
     aaa[bC]BcBc
     aaa[bb]BcBc
     aaab[bB]cBc
     aaab[bb]cBc
     aaabbb[cB]c
     aaabbb[cc]c
     ```

6. **Derivação Final**:
   - **Resultado**:
     ```plaintext
     aaabbbccc
     ```
   - Este processo envolve a expansão repetitiva e a transformação sensível ao contexto, onde cada $B$ se expande em mais 'b's e cada $C$ se expande em mais 'c's para igualar o número de 'a's.

Este método garante que a quantidade de 'a's, 'b's, e 'c's aumente de maneira sincronizada, cumprindo a condição da linguagem $\{a^n b^n c^n | n \geq 1\}$.
- Na gramática fornecida, regras como $bB \rightarrow bb$ e $bC \rightarrow bc$ modificam o contexto ao redor das variáveis, adicionando terminais diretamente adjacentes à variável que está sendo substituída, o que é característico das gramáticas sensíveis ao contexto.
- Isso significa que a substituição de uma variável pode depender dos símbolos à sua esquerda e direita, ou seja, as regras podem alterar uma parte da string somente baseando-se nos símbolos que a cercam, não apenas com base na variável isolada (diferentemente das gramáticas livres de contexto)

Esses conceitos refletem a natureza restritiva, mas ainda bastante poderosa, das linguagens sensíveis ao contexto, situando-as de forma única na hierarquia de Chomsky.

## Slide 21 - Máquina de Turing com Fita Limitada

Uma **Máquina de Turing com Fita Limitada** é uma variação da máquina de Turing padrão, onde a fita é restrita ao tamanho da palavra de entrada mais duas células adicionais para marcadores de início e fim.

**Configuração Formal da Máquina**

- **Formalmente definida como uma 8-tupla**:
  - $M = \{\Sigma, Q, \delta, q_0, F, V, \circledast, \dagger\}$

**Componentes**

- $\Sigma$: Alfabeto de entrada, contendo os símbolos que a máquina pode processar.
- $Q$: Conjunto finito de estados da máquina.
- $\delta$: Função de transição, definida como:
  $$\delta: Q \times (\Sigma \cup V \cup \{\circledast, \dagger\}) \rightarrow 2^{Q \times (\Sigma \cup V \cup \{\circledast, \dagger\}) \times \{E, D\}}$$
  - Esta função determina as ações da máquina, incluindo movimentos da cabeça de leitura/gravação (Esquerda - $E$, Direita - $D$), as operações sobre os símbolos na fita e as mudanças de estados
- $q_0$: Estado inicial da máquina.
- $F$: Conjunto de estados finais, onde a máquina pode parar e aceitar a entrada.
- $V$: Alfabeto auxiliar, usado para símbolos internos que podem não ser parte do alfabeto de entrada.
- $\circledast$: Símbolo especial que marca o início da fita.
- $\dagger$: Símbolo especial que marca o fim da fita.

**Propósito**

Esta configuração é especialmente útil para lidar com linguagens sensíveis ao contexto, onde as restrições de tamanho da fita simulam as restrições contextuais das produções em gramáticas sensíveis ao contexto.

- Exploração de Limitações Computacionais
    - Simulação de Restrições Contextuais
        - Ao limitar o tamanho da fita ao tamanho da entrada mais um espaço fixo para marcadores especiais, esta máquina simula um ambiente onde as operações não podem extrapolar um contexto delimitado
        - Isso é crucial para estudar como certas linguagens sensíveis ao contexto podem ser manipuladas ou reconhecidas dentro de limites estritos.
    - Eficiência e Complexidade
        - A restrição de tamanho da fita também permite explorar a eficiência computacional e complexidade de problemas que podem ou não ser solucionados dentro de limites espaciais fixos
        - Por exemplo, algumas tarefas que são decidíveis em uma Máquina de Turing padrão podem se tornar indecidíveis quando submetidas a limitações de espaço.
    - Desenvolvimento de Algoritmos
        - Em contextos práticos, entender as limitações impostas por tais máquinas pode levar ao desenvolvimento de algoritmos mais eficientes que operam dentro de restrições de espaço fixo
        - Por exemplo, algoritmos que irão operar em dispositivos embutidos e sistemas com recursos limitados
    - Modelagem de Sistemas Confinados

A linguagem $L = \{ ww | w \text{ é de } \{a, b\}^*\}$ é interessante porque cada palavra na linguagem consiste em uma repetição exata de uma sequência de caracteres. Abaixo, descrevemos como uma Máquina de Turing com fita limitada pode reconhecer essa linguagem.

**Descrição da Máquina**

Esta Máquina de Turing utiliza uma estratégia especial para verificar se a palavra na fita é uma duplicação exata de uma sequência de caracteres. A ideia é que a máquina leia a primeira metade da palavra, marcando os caracteres lidos e depois verificar se a segunda metade corresponde exatamente aos caracteres marcados.

**Implementação**

A máquina começa no estado inicial onde lê o primeiro caractere, o marca de alguma forma (por exemplo, substituindo 'a' ou 'b' por 'X'), e então move para a direita. Quando atinge o meio da palavra (detectado por um marcador especial ou pelo fim dos caracteres marcados), ela começa a comparar os caracteres subsequentes com os marcados. Se todos corresponderem até o final da palavra, a entrada é aceita; caso contrário, é rejeitada.

### Configurações da Máquina de Turing de Fita Limitada para..
$L = \{ ww | w \text{ é de } \{a, b\}^*\}$

- **Formalmente definida como uma 8-tupla**:
  - $M = \{\Sigma, Q, \delta, q_0, F, V, \circledast, \dagger\}$

**Componentes**

- $\Sigma: \{a, b\}$
- $Q: \{q_0, q_1, q_2, q_3, q_4, q_5, q_6, q_7, q_8, q_9, q_{10}, q_{11}, q_f\}$
- $q_0: \{q_0\}$
- $F: \{q_f\}$
- $V: \{X, Y\}$
- $\circledast$: Símbolo especial que marca o início da fita.
- $\dagger$: Símbolo especial que marca o fim da fita.
- $\delta$: Função de transição, definida como:
  $$\delta: Q \times (\Sigma \cup V \cup \{\circledast, \dagger\}) \rightarrow 2^{Q \times (\Sigma \cup V \cup \{\circledast, \dagger\}) \times \{E, D\}}$$  
tal que,  
- $\delta(q_0)$
   - $(q_0, \circledast) \to (q_1, \circledast, D)$

- $\delta(q_1)$
   - $(q_1, a) \to (q_2, X, D)$
   - $(q_1, b) \to (q_3, X, D)$
   - $(q_1, \dagger) \to (q_f, \dagger, E)$

- $\delta(q_2)$ (Não-determinístico ao ler 'a')
   - $(q_2, a) \to (q_2, a, D)$  # Continua lendo 'a's
   - $(q_2, a) \to (q_4, Y, E)$  # Marca e move para a esquerda após processar 'a'
   - $(q_2, b) \to (q_2, b, D)$  # Continua lendo 'b's

- $\delta(q_3)$ (Não-determinístico ao ler 'b')
   - $(q_3, a) \to (q_3, a, D)$  # Continua lendo 'a's
   - $(q_3, b) \to (q_3, b, D)$  # Continua lendo 'b's e se mantém em $q_3$
   - $(q_3, b) \to (q_4, Y, E)$  # Marca 'b' como 'Y' e move para a esquerda para $q_4$

- $\delta(q_4)$
   - $(q_4, a) \to (q_4, a, E)$  # Move para a esquerda, relendo 'a'
   - $(q_4, b) \to (q_4, b, E)$  # Move para a esquerda, relendo 'b'
   - $(q_4, X) \to (q_5, X, D)$  # Move para a direita após verificar 'X'

   - $(q_6, a) \to (q_6, a, D)$: Ao ler 'a', mantém 'a' e move para a direita, permanece em $q_6$.
   - $(q_6, b) \to (q_6, b, D)$: Ao ler 'b', mantém 'b' e move para a direita, permanece em $q_6$.
   - $(q_6, Y) \to (q_7, Y, D)$: Ao ler 'Y', mantém 'Y' e move para a direita, transita para $q_7$.

- $\delta(q_8)$
   - $(q_8, a) \to (q_8, a, D)$: Ao ler 'a', mantém 'a' e move para a direita, permanece em $q_8$.
   - $(q_8, b) \to (q_8, b, D)$: Ao ler 'b', mantém 'b' e move para a direita, permanece em $q_8$.
   - $(q_8, Y) \to (q_9, Y, D)$: Ao ler 'Y', mantém 'Y' e move para a direita, transita para $q_9$.

- $\delta(q_7)$:
  - $(q_7, Y) \to (q_7, Y, D)$: Ao ler 'Y', mantém 'Y' e move para a direita, permanece em $q_7$.
  - $(q_7, a) \to (q_10, Y, E)$: Ao ler 'a', substitui por 'Y' e move para a esquerda, transita para $q_10$.

- $\delta(q_9)$:
  - $(q_9, Y) \to (q_9, Y, D)$: Ao ler 'Y', mantém 'Y' e move para a direita, permanece em $q_9$.
  - $(q_9, b) \to (q_10, Y, E)$: Ao ler 'b', substitui por 'Y' e move para a esquerda, transita para $q_10$.

- $\delta(q_{10})$:
  - $(q_{10}, Y) \to (q_{10}, Y, E)$: Ao ler 'Y', mantém 'Y' e move para a esquerda, permanece em $q_{10}$.
  - $(q_{10}, a) \to (q_{10}, a, E)$: Ao ler 'a', mantém 'a' e move para a esquerda, permanece em $q_{10}$.
  - $(q_{10}, b) \to (q_{10}, b, E)$: Ao ler 'b', mantém 'b' e move para a esquerda, permanece em $q_{10}$.

- $\delta(q_{11})$:

  - $(q_{11}, \dagger) \to (q_f, \dagger, E)$: Ao ler '$\dagger$', mantém '$\dagger$' e move para a esquerda, transita para o estado final $q_f$

### Configuração Inicial da Máquina de Turing

- **Fita inicial**: $\circledast abab \dagger$
- **Cabeça de leitura**: Posicionada no $\circledast$
- **Estado inicial**: $q_0$

### Funcionamento Detalhado

1. **Passo 1**: No estado $q_0$, a cabeça lê o $\circledast$, escreve $\circledast$, e move para a direita, mudando para o estado $q_1$.

   - **Fita**: $\circledast \underline{a}bab \dagger$

2. **Passo 2**: Em $q_1$, lê 'a', substitui por 'X' e move para a direita, mudando para o estado $q_2$.
   - **Mudança de Estado**: $q_1 \to q_2$
   - **Fita**: $\circledast X\underline{b}ab \dagger$

4. **Passo 4**: Em $q_2$, lê 'a', substitui por 'Y' e move para a esquerda.
   - **Mudança de Estado**: $q_2 \to q_4$

   - **Fita**: $\circledast \underline{X}bYb \dagger$

4. **Passo 6**: Em $q_4$, lê 'X', mantém 'X' e move para a direita.
   - **Mudança de Estado**: $q_4 \to q_5$
   - **Fita**: $\circledast X\underline{b}Yb \dagger$

### Resultado

A máquina termina no estado $q_f$ com a fita mostrando $\circledast XXYY \dagger$, indicando que a palavra de entrada foi duplicada corretamente segundo a linguagem especificada $L = \{ ww | w \text{ é de } \{a, b\}^*\}$.

````python
from collections import deque
# EM CONSTRUÇÃO....
   #- $(q_2, a) \to (q_2, a, D)$  # Continua lendo 'a's
   #- $(q_2, a) \to (q_4, Y, E)$  # Marca e move para a esquerda após processar 'a'
   #- $(q_2, b) \to (q_2, b, D)$  # Continua lendo 'b's
class TuringMachine:
    def __init__(self, tape, initial_state):
         # Convertemos a string de entrada em uma lista para permitir modificações
        self.tape = list(tape)
        self.head_position = 0
        self.current_state = initial_state

        # Dicionário de transições
        # Chave: (estado_atual, símbolo_lido)
        # Valor: conjunto de (novo_estado, símbolo_escrito, direção)
        self.states = {
            # q0
            ('q0', '0'):{('q1', '0', 'D')},

            # q1
            ('q1','a'): {('q2','X','D')},           # Lê 'a', escreve 'X', move p/ direita -> q2
            ('q1','b'): {('q3','X','D')},           # Lê 'b', escreve 'X', move p/ direita -> q3
            ('q1','t'): {('qf','t','E')},           # Lê 't', aceita (vai p/ qf)

            # q2
            # Duas transições possíveis para símbolo a:
            #  - (a,a,D) significa "continue lendo e avançando"
            #  - (a,Y,E) significa "substitui por Y e volta para a esquerda"
            ('q2','a'): {('q2','a','D'), ('q4','Y','E')},
            ('q2','b'): {('q2','b','D')},

            # q3 (simétrico a q2, só que ao ler 'b')
            ('q3','a'): {('q3','a','D')},
            ('q3','b'): {('q3','b','D'), ('q4','Y','E')},

            # q4
            # Loop para “apagar” a ou b voltando para a esquerda;
            # se encontrar X, volta a andar para a direita indo a q5
            ('q4','a'): {('q4','a','E')},
            ('q4','b'): {('q4','b','E')},
            ('q4','X'): {('q5','X','D')},

            # q5
            # Aqui começa outro “ciclo” de verificação.
            # Exemplos de transições possíveis:
            ('q5','a'): {('q6','X','D')},
            ('q5','b'): {('q8','X','D')},
            ('q5','Y'): {('q11','Y','D')}, # se encontrar Y, pula para q11 (depende do diagrama)

            # q6
            ('q6','a'): {('q6','a','D')},
            ('q6','b'): {('q6','b','D')},
            ('q6','Y'): {('q7','Y','D')},

            # q7
            ('q7','Y'): {('q7','Y','D')},
            ('q7','a'): {('q10','Y','E')},
            
            # q8 (simétrico a q6)
            ('q8','a'): {('q8','a','D')},
            ('q8','b'): {('q8','b','D')},
            ('q8','Y'): {('q9','Y','D')},

            # q9 (simétrico a q7)
            ('q9','Y'): {('q9','Y','D')},
            ('q9','b'): {('q10','Y','E')},

            # q10
            ('q10','a'): {('q10','a','E')},
            ('q10','b'): {('q10','b','E')},
            ('q10','Y'): {('q10','Y','E')},
            ('q10','X'): {('q5','X','D')},  # retorna para q5 e recomeça ciclo

            # q11
            ('q11','Y'): {('q11','Y','D')},
            ('q11','t'): {('qf','t','E')}
        }
        self.final_states = ['qf']

    def possible_actions(self, tape, head, state):
        """
        Retorna todas as transições possíveis para (state, simbolo_lido).
        Caso não exista entrada no dicionário, retorna um conjunto vazio.
        """
        if head < 0 or head >= len(tape):
            return set()  # Cabeça fora da fita limitada
        symbol = tape[head]
        return self.states.get((state, symbol), set())

    def run(self):
        """
        Usa uma fila (BFS) para simular todas as configurações possíveis
        de forma não determinística.
        """
        # Fila de configurações: cada elemento é (tape_atual, head_atual, state_atual)
        queue = deque()
        visited = set()

        # Cria a configuração inicial e marca como visitada
        initial_config = (tuple(self.tape), self.head_position, self.current_state)
        queue.append(initial_config)
        visited.add(initial_config)

        while queue:
            current_tape, current_head, current_state = queue.popleft()

            # Imprimir estado atual (debug)
            print("Current state:", current_state)
            print("Tape:", ''.join(current_tape))
            print("Head position:", current_head)
            print("-------------------------")

            # Verifica se chegou a um estado final
            if current_state in self.final_states:
                print("Machine halts at final state:", current_state)
                print("Final tape configuration:", ''.join(current_tape))
                return # Encerra ao encontrar o primeiro estado final

            # Obtém as transições possíveis para o par (current_state, symbol_lido)
            actions = self.possible_actions(current_tape, current_head, current_state)
            
            if not actions:
                # Sem transições, a máquina para nesse ramo
                print(f"Nenhuma transição para ({current_state}, '{current_tape[current_head]}'). Caminho encerra.")
                continue

            # Explorar todas as transições (não determinismo)
            for (new_state, new_char, direction) in actions:
                # Criar nova fita (como lista) a partir da fita atual (tupla)
                new_tape_list = list(current_tape)
                # Escrever novo símbolo
                new_tape_list[current_head] = new_char

                # Ajusta a cabeça
                if direction == 'D':
                    new_head = current_head + 1
                else:  # 'E'
                    new_head = current_head - 1

                # Verifica se ainda estamos dentro dos limites da fita
                if 0 <= new_head < len(new_tape_list):
                    # Criar nova configuração
                    new_config = (tuple(new_tape_list), new_head, new_state)
                    
                    # Se ainda não visitamos essa configuração, adicionamos na fila
                    if new_config not in visited:
                        visited.add(new_config)
                        queue.append(new_config)

    def run_path(self):
        """
        Usa uma fila (BFS) para simular todas as configurações possíveis
        de forma não determinística, armazenando o caminho de estados percorrido.
        """

        # Fila de configurações:
        # cada elemento é ((tape, head, state), path)
        queue = deque()
        visited = set()

        # Contador global para indexar cada caminho/ramo
        path_id_counter = 0

        # Configuração inicial
        initial_config = (tuple(self.tape), self.head_position, self.current_state)
        initial_path = [self.current_state]  # lista com o estado inicial
        
        # Enfileirar a configuração inicial com path_id = 0
        queue.append((initial_config, initial_path, path_id_counter))
        visited.add(initial_config)

        # Incrementa para que o próximo ramo tenha outro ID
        path_id_counter += 1

        while queue:
            # Desempilha a configuração atual e o caminho
            (current_tape, current_head, current_state), path, curr_path_id = queue.popleft()

            # Mostra informações de debug
            print(f"[Caminho #{curr_path_id}]")
            print("Current state:", current_state)
            print("Tape:", ''.join(current_tape))
            print("Head position:", current_head)
            print("Path so far:", " -> ".join(path))
            print("-------------------------")

            # Verifica se chegou a um estado final
            if current_state in self.final_states:
                print(f"Machine halts at final state: {current_state}")
                print(f"Caminho #{curr_path_id} concluído.")
                print("Path of states:", " -> ".join(path))
                print("Final tape configuration:", ''.join(current_tape))
                return  # Encerra ao encontrar um estado final (ou comente se quiser continuar)


            # Obtém as transições possíveis
            actions = self.possible_actions(current_tape, current_head, current_state)

            if not actions:
                # Sem transições, esse ramo do não-determinismo encerra aqui
                print(f"Nenhuma transição para ({current_state}, '{current_tape[current_head]}'). "
                    f"Caminho #{curr_path_id} encerrado: {' -> '.join(path)}")
                continue

            # Para cada transição (não determinismo)
            for (new_state, new_char, direction) in actions:
                # Criar nova fita como lista para podermos editar
                new_tape_list = list(current_tape)
                new_tape_list[current_head] = new_char  # escreve novo símbolo

                # Ajustar cabeça
                if direction == 'D':
                    new_head = current_head + 1
                else:  # 'E'
                    new_head = current_head - 1

                # Verificar se o movimento permanece dentro dos limites
                if 0 <= new_head < len(new_tape_list):
                    new_config = (tuple(new_tape_list), new_head, new_state)

                    # Só exploramos se ainda não visitamos
                    if new_config not in visited:
                        visited.add(new_config)

                        # Atualiza o caminho de estados
                        new_path = path + [new_state]

                        # Enfileirar a nova configuração + o novo caminho
                        queue.append((new_config, new_path, path_id_counter))
                        path_id_counter += 1
                else:
                    print("Movimento fora dos limites da fita. Transição descartada.")

        # Caso chegue aqui, significa que não existem mais configurações na fila
        # e nenhum estado final foi alcançado.
        print("Nenhum estado final foi alcançado ao esgotar todas as configurações possíveis.")

# Example usage:
tm = TuringMachine("0ababt", 'q0')
tm.run_path()
````
